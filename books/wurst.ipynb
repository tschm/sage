```python
S=create_S3(matrix=T, digits=3)
S
```

```python
K = 20  # extrapolation
P = 30  # power cutoff
Ignore = -22
Psi = np.zeros((K,T.shape[0],P))

# compute the sums Psi_{i} explicitly for up to 5 digits
for k in range(1, P):
    for index, s in np.ndenumerate(S):
        Psi[index[0],index[1],k] = psi(s, k, prec=20);
    end
end

Psi[:,:,1]
```
