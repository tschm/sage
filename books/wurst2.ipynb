```python
import numpy as np
from kempner.util import  create_S3, create_f, psi, matrix_42, Psi_matrix, aCoeff
```

```python
T = matrix_42()
T
```

```python
f, A, B = create_f(T)
B
```

```python
S=create_S3(matrix=T, digits=3)
```

```python
Psi = Psi_matrix(S)
```

```python
# forward interpolate rows of Psi
for i in range(S.shape[0], Psi.shape[0]):
    for k in range(1, Psi.shape[2]):
        #if Psi[i-1,1,k]>0:
        for m in range(0, T.shape[1]):
            [J,L] = np.where(f[:,:,m]);
            for l,j in zip(L,J):
                for w in range(0, Psi.shape[2]-k):
                    Psi[i,j,k] = Psi[i,j,k]+aCoeff(k,w,m)*Psi[i-1,l,k+w]                    
```
